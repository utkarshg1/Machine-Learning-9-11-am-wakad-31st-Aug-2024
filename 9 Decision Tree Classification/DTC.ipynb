# Decision Tree Classification

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1 - Read The dataset

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

### Step 2 - Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 - Seperate X and Y(species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4 - Perform preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
num_pipe
```

### Step 5 - Train Test Split
test_size = 0.33, random_state=42

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.33, random_state=42
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Step 6 - Model Building

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="gini"
)
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

### Tuning the model

```python
params = {
    "max_depth": [1, 2, 3, 4, 5],
    "min_samples_split": [4, 5, 6, 7, 8],
    "min_samples_leaf": [4, 5, 6, 7, 8],
    "criterion": ["gini", "entropy"]
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring="f1_macro")
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
```

```python
best_dtc
```

### Step 7 - Evaluate the model

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
ypred_test = best_dtc.predict(xtest)
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest, ypred_test))
```

### This is a good model because f1_score > 0.8 for both train and test

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
plot_tree(
    best_dtc, 
    feature_names=xtrain.columns, 
    class_names= best_dtc.classes_,
    filled=True
)
plt.show()
```
